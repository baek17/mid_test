#### 1.4.4  Smoothing Images

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img=cv2.imread('IMG_6504.JPG')
rows,cols,ch = img.shape    # 이미지의 차원(행, 열, 채널)을 가져옴
pts1 = np.float32([[170,190],[760,196],[808,1096],[100,1080]])  # 입력 이미지에서 변환할 포인트
pts2 = np.float32([[0,0],[400,0],[400,600],[0,600]])    # 출력 이미지에서 대응되는 포인트
M =  cv2.getPerspectiveTransform(pts1,pts2) # 변환 행렬을 계산
dst = cv2.warpPerspective(img,M,(400,600))  # 이미지에 원근 변환을 적용
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
```

```python
img=cv2.imread('dsu2.jpg')
rows,cols,ch = img.shape
pts1 = np.float32([[180,240],[820,220],[820,500],[180,500]])
pts2 = np.float32([[0,0],[820,0],[820,500],[0,600]])
M =  cv2.getPerspectiveTransform(pts1,pts2)
dst = cv2.warpPerspective(img,M,(830,510))
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
```
