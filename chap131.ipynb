#### Chap 1.3.1 Basic Operations on Images

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('cat.jpg')
```

```python
plt.imshow(img, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([]) # X, Y 축 값 숨김
plt.show()
```

```python
cv2.namedWindow('cat', cv2.WINDOW_NORMAL)
cv2.imshow('cat',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imwrite('messigray.png',img)
```

```python
img.shape
# 배열의 모양을 반환
# 이미지가 배열로 저장되어 있을 때,
# shape 속성은 (height, width, channels) 형식으로 이미지의 높이, 너비, 그리고 색상 채널(BGR) 수를 나타냄
```

```python
img[100, 100]
# 이미지의 (100, 100) 좌표에 있는 픽셀 값을 출력, BGR 형식의 값이 반환
```

```python
img[100, 100, 0] = 0
# 해당 좌표의 파란색(B) 채널 값을 0으로 수정
```

```python
img[100, 100]
# 변경된 값을 확인
```

```python
cv2.namedWindow('cat', cv2.WINDOW_NORMAL)
cv2.imshow('cat',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imwrite('messigray.png',img)
```

```python
img[100, 100] = [0, 0, 0]
# (100, 100) 좌표에 있는 픽셀 값을 검정색([0, 0, 0])으로 설정
# (100, 100) 위치에 있는 픽셀이 원래의 색상에서 검정색으로 변경됨
```

```python
cv2.namedWindow('cat', cv2.WINDOW_NORMAL)
cv2.imshow('cat',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imwrite('messigray.png',img)
```

```python
img[0:10, 0:10]
# img 배열의 특정 영역을 슬라이스(slice)하여 반환
# 이미지의 (0, 0)부터 (10, 10)까지의 부분을 선택
```

```python
img[0:10, 0:10] = [0, 0, 0]
# (0, 0)부터 (10, 10)까지의 좌표에 있는 픽셀을 검정색([0, 0, 0])으로 설정
```

```python
cv2.namedWindow('cat', cv2.WINDOW_NORMAL)
cv2.imshow('cat',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imwrite('messigray.png',img)
```

```python
img[600:800, 600:1000, 0] = 0
# (600, 800)부터 (1000, 1000)까지의 파란색 채널 값을 0으로 설정
# 해당 영역에서 파란색이 제거
```

```python
cv2.namedWindow('cat', cv2.WINDOW_NORMAL)
cv2.imshow('cat',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imwrite('messigray.png',img)
```

```python
plt_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(plt_img, interpolation = 'bicubic')
plt.show()
```

#### 1.3.1 실습

```python
import cv2
import numpy as np
img = cv2.imread('cat.jpg')
```

```python
px = img[100,100]   # (100, 100) 좌표의 픽셀 값 접근
print (px)  # 해당 좌표 픽셀 값을 출력
```

```python
blue = img[100,100,0]    # (100, 100) 픽셀의 Blue 채널 값 접근
print (blue)
```

```python
img[100,100] = [255,255,255]    # (100, 100) 픽셀 값을 흰색으로 변경
print (img[100,100])    # 변경된 값을 출력
```

```python
img.item(10,10,2)   # (10, 10) 좌표의 Red 채널 값 접근
```

```python
img.itemset((10,10,2),100)  # (10, 10) 좌표의 Red 채널 값을 100으로 설정
img.item(10,10,2)   # 다시 값을 출력
```

```python
print (img.shape)    # 이미지의 차원(높이, 너비, 채널 수) 출력
```

```python
print (img.size)    # 이미지의 총 픽셀 수 출력
```

```python
print (img.dtype)   # 이미지 데이터 타입 출력
```

```python
ball = img[280:340, 330:390]
# (280, 340)에서 (330, 390)까지 영역을 잘라서 ball에 저장
img[273:333, 100:160] = ball
# 잘라낸 영역을 (273, 333)에서 (100, 160) 좌표로 붙여넣음
```

```python
b,g,r = cv2.split(img)  # B, G, R 채널을 각각 분리
img = cv2.merge((b,g,r))    # 다시 B, G, R 채널을 합성
```

```python
b = img[:,:,0]   # Blue 채널을 가져옴
```

```python
img[:,:,2] = 0  # Red 채널 값을 모두 0으로 설정 (이미지에서 Red 색상 제거)
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
BLUE = [255,0,0]

img1 = cv2.imread('opencv_logo.png')
# cv2.copyMakeBorder()는 이미지에 테두리를 추가하는 함수
# BORDER_* 옵션은 테두리를 어떻게 추가할지를 결정
replicate = cv2.copyMakeBorder(img1,10,10,10,10,cv2.BORDER_REPLICATE)   # 가장자리 픽셀을 복사
reflect = cv2.copyMakeBorder(img1,10,10,10,10,cv2.BORDER_REFLECT)   # 픽셀을 반사
reflect101 = cv2.copyMakeBorder(img1,10,10,10,10,cv2.BORDER_REFLECT_101)    # 반사를 하되 마지막 픽셀을 제외하고 반사
wrap = cv2.copyMakeBorder(img1,10,10,10,10,cv2.BORDER_WRAP) # 이미지가 반복되도록 테두리 설정
constant= cv2.copyMakeBorder(img1,10,10,10,10,cv2.BORDER_CONSTANT,value=BLUE)   # 지정한 색상으로 테두리 추가 (여기서는 BLUE)

plt.subplot(231),plt.imshow(img1,'gray'),plt.title('ORIGINAL')
plt.subplot(232),plt.imshow(replicate,'gray'),plt.title('REPLICATE')
plt.subplot(233),plt.imshow(reflect,'gray'),plt.title('REFLECT')
plt.subplot(234),plt.imshow(reflect101,'gray'),plt.title('REFLECT_101')
plt.subplot(235),plt.imshow(wrap,'gray'),plt.title('WRAP')
plt.subplot(236),plt.imshow(constant,'gray'),plt.title('CONSTANT')
plt.show()
```
