#### 1.4.5  Morphological Transformations

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('j.png',0)
kernel = np.ones((3, 3),np.uint8)
# 형태학적 연산을 위한 커널(구조 요소) 정의
erosion = cv2.erode(img, kernel, iterations = 2)
# 이미지에 침식(erosion) 연산 수행
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(erosion),plt.title('Output')
```

```python
img = cv2.imread('j.png',0)
kernel = np.ones((5,5),np.uint8)
# 팽창(dilation)을 위한 더 큰 커널 정의
dilation = cv2.dilate(img,kernel,iterations = 1)
# 이미지에 팽창 연산 수행
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dilation),plt.title('Output')
```

```python
img = cv2.imread('j2.png',0)
kernel = np.ones((5,5),np.uint8)
# 형태학적 개방을 위한 커널 정의
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
# 이미지에 개방 연산 수행
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(opening),plt.title('Output')
```

```python
img = cv2.imread('j3.png',0)
kernel = np.ones((5,5),np.uint8)
# 형태학적 폐쇄를 위한 커널 정의
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
# 이미지에 폐쇄 연산 수행
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(closing),plt.title('Output')
```

```python
img = cv2.imread('j.png',0)
kernel = np.ones((5,5),np.uint8)
# 형태학적 그래디언트(gradient)를 위한 커널 정의
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)
# 이미지에 그래디언트 연산 수행
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(gradient),plt.title('Output')
```

```python
img = cv2.imread('j.png',0)
kernel = np.ones((5,5),np.uint8)
# 토패트(tophat) 연산을 위한 커널 정의
tophat = cv2.morphologyEx(img, cv2.MORPH_TOPHAT, kernel)
# 이미지에 토패트 연산 수행
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(tophat),plt.title('Output')
```

```python
img = cv2.imread('j.png',0)
kernel = np.ones((5, 5),np.uint8)
# 블랙햇(blackhat) 연산을 위한 커널 정의
blackhat = cv2.morphologyEx(img, cv2.MORPH_BLACKHAT, kernel)
# 이미지에 블랙햇 연산 수행
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(blackhat),plt.title('Output')
```
