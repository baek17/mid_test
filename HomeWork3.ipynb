```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img=cv2.imread('dsu2.jpg')
rows,cols,ch = img.shape    # 이미지의 크기(행, 열, 채널 수)를 변수에 저장
pts1 = np.float32([[180,240],[820,220],[820,500],[180,500]])
# 원본 이미지에서 변환할 사각형의 4개 모서리 좌표를 설정, 이 좌표는 원본 이미지에서 변환할 영역을 정의
pts2 = np.float32([[0,0],[820,0],[820,500],[0,600]])
# 변환된 이미지에서의 4개 모서리 좌표를 설정, 이 좌표는 변환 후에 원하는 형태의 사각형을 정의
M =  cv2.getPerspectiveTransform(pts1,pts2) # 두 집합의 좌표(pts1, pts2) 간의 원근 변환을 위한 변환 행렬 M을 계산
dst = cv2.warpPerspective(img,M,(830,510))  # 원본 이미지에 계산된 변환 행렬 M을 적용합니다. (830, 510)은 출력 이미지의 크기를 지정
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
```
