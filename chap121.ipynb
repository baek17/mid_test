```python
import numpy as np
import cv2
```

```python
# 기본적으로 이미지를 컬러로 읽음
img = cv2.imread('cat.jpg')
```

```python
cv2.imshow('cat',img)   # 읽은 이미지를 'cat'이라는 윈도우 창에 표시
cv2.waitKey(0)  # 키 입력을 기다림, 0이면 무한정 기다림
cv2.destroyAllWindows() # 열려 있는 모든 윈도우를 닫음
```

```python
cv2.namedWindow('cat', cv2.WINDOW_NORMAL)   # 창 생성, cv2.WINDOW_NORMAL 사용하면 창 크기 조절 가능
cv2.imshow('cat',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imwrite('messigray.png',img)    # 읽은 이미지를 'messigray.png' 파일로 저장
```

```python
img = cv2.imread('cat.jpg',0)   # cat.jpg' 이미지를 그레이스케일로 읽음, 0은 그레이스케일 모드를 의미
cv2.imshow('image',img) # 'image'라는 창에 그레이스케일 이미지 표시
k = cv2.waitKey(0)
if k == 27: # k == 27은 ESC 키가 눌렸을 때 창을 닫음
    cv2.destroyAllWindows()
elif k == ord('s'): # 's' 키가 눌렸을 때 이미지를
    cv2.imwrite('catgray00.png',img)    # catgray00.png'로 저장하고
cv2.destroyAllWindows() # 창을 닫음
```
